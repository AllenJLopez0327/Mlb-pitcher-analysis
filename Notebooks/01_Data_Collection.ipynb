### Exploratory Data Analysis

```python
import pandas as pd
import os

# Load all raw datasets
data_path = '../Data/Raw/'

# FanGraphs datasets
fg_stats = pd.read_csv(data_path + 'fangraphs-Stats.csv')
fg_pitch_type = pd.read_csv(data_path + 'fangraphs-Pitch Type.csv')
fg_velocity = pd.read_csv(data_path + 'fangraphs-Velocity.csv')
fg_h_movement = pd.read_csv(data_path + 'fangraphs-H-Movement.csv')
fg_v_movement = pd.read_csv(data_path + 'fangraphs-V-Movement.csv')
fg_spin = pd.read_csv(data_path + 'fangraphs-Spin.csv')
fg_arm_angle = pd.read_csv(data_path + 'fangraphs-Arm Angel.csv')
fg_edge_cases = pd.read_csv(data_path + 'Edge Cases Relievers.csv')

# Cy Young voting data
cy_young = pd.read_csv(data_path + 'CY young data.csv')

# Print shapes to verify
datasets = {
    'fg_stats': fg_stats,
    'fg_pitch_type': fg_pitch_type,
    'fg_velocity': fg_velocity,
    'fg_h_movement': fg_h_movement,
    'fg_v_movement': fg_v_movement,
    'fg_spin': fg_spin,
    'fg_arm_angle': fg_arm_angle,
    'fg_edge_cases': fg_edge_cases,
    'cy_young': cy_young
}

for name, df in datasets.items():
    print(f'{name}: {df.shape} — columns: {list(df.columns[:5])}')
```

```python
for name, df in datasets.items():
    print(f'\n{"="*60}')
    print(f'{name}')
    print(f'{"="*60}')
    print(df.head(3).to_string())
```
