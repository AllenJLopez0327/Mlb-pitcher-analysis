## Loading Data 

```python
import pandas as pd
import os

# Processed datasets
processed_path = '../Data/Processed/'

fg_stats = pd.read_csv(processed_path + 'fangraphs-Stats.csv')
fg_pitch_type = pd.read_csv(processed_path + 'fangraphs-Pitch Type.csv')
fg_velocity = pd.read_csv(processed_path + 'fangraphs-Velocity.csv')
fg_h_movement = pd.read_csv(processed_path + 'fangraphs-H-Movement.csv')
fg_v_movement = pd.read_csv(processed_path + 'fangraphs-V-Movement.csv')
fg_spin = pd.read_csv(processed_path + 'fangraphs-Spin.csv')
fg_advanced = pd.read_csv(processed_path + 'Fangraphs-Advanced.csv')
cy_young = pd.read_csv(processed_path + 'CY-young.csv')

# Print shapes to verify
datasets = {
    'fg_stats': fg_stats,
    'fg_pitch_type': fg_pitch_type,
    'fg_velocity': fg_velocity,
    'fg_h_movement': fg_h_movement,
    'fg_v_movement': fg_v_movement,
    'fg_spin': fg_spin,
    'fg_advanced': fg_advanced,
    'cy_young': cy_young
}

for name, df in datasets.items():
    print(f'{name}: {df.shape} — columns: {list(df.columns)}')

for name, df in datasets.items():
    print(f'{name}: {df.shape} — columns: {list(df.columns[:5])}')
```

```python
for name, df in datasets.items():
    print(f'\n{"="*60}')
    print(f'{name}')
    print(f'{"="*60}')
    print(df.head(3).to_string())
```

## Cleaning the Datasets

### Cleaned datasets apart from cy young

```python
# Clean all FanGraphs files - columns already cleaned in Excel
# Just need to standardize Team column

fg_stats_clean = fg_stats.copy()
fg_stats_clean['Team'] = fg_stats_clean['Team'].replace('- - -', 'TOT')
fg_stats_clean = fg_stats_clean.rename(columns={'vFA (pi)': 'vFA'})

fg_pitch_type_clean = fg_pitch_type.copy()
fg_pitch_type_clean['Team'] = fg_pitch_type_clean['Team'].replace('- - -', 'TOT')
pitch_cols = [c for c in fg_pitch_type_clean.columns if c.endswith('%')]
fg_pitch_type_clean[pitch_cols] = fg_pitch_type_clean[pitch_cols].fillna(0)

fg_velocity_clean = fg_velocity.copy()
fg_velocity_clean['Team'] = fg_velocity_clean['Team'].replace('- - -', 'TOT')

fg_h_movement_clean = fg_h_movement.copy()
fg_h_movement_clean['Team'] = fg_h_movement_clean['Team'].replace('- - -', 'TOT')

fg_v_movement_clean = fg_v_movement.copy()
fg_v_movement_clean['Team'] = fg_v_movement_clean['Team'].replace('- - -', 'TOT')

fg_spin_clean = fg_spin.copy()
fg_spin_clean['Team'] = fg_spin_clean['Team'].replace('- - -', 'TOT')

fg_advanced_clean = fg_advanced.copy()
fg_advanced_clean['Team'] = fg_advanced_clean['Team'].replace('- - -', 'TOT')

# Verify all
for name, df in [('fg_stats', fg_stats_clean), 
                  ('fg_pitch_type', fg_pitch_type_clean),
                  ('fg_velocity', fg_velocity_clean), 
                  ('fg_h_movement', fg_h_movement_clean),
                  ('fg_v_movement', fg_v_movement_clean), 
                  ('fg_spin', fg_spin_clean),
                  ('fg_advanced', fg_advanced_clean)]:
    print(f'{name}: {df.shape} — nulls: {df.isnull().sum().sum()}')
```

### Cy young dataset

```python
print(cy_young.shape)
print(cy_young.dtypes)
print()
print(cy_young.head(6).to_string())
```

```python
# Clean cy_young
cy_young_clean = cy_young.copy()

# Rename for consistency
cy_young_clean = cy_young_clean.rename(columns={'Year': 'Season', 'Vote Pts': 'Vote_Pts', '1st Place': '1st_Place'})

# Convert Share to float
cy_young_clean['Share'] = cy_young_clean['Share'].str.replace('%', '').astype(float) / 100

# Add winner column
cy_young_clean['CY_winner'] = (cy_young_clean['Rank'] == 1).astype(int)

# Handle multi-team
cy_young_clean['Team'] = cy_young_clean['Team'].replace('TOT', 'TOT')

print(cy_young_clean.shape)
print(cy_young_clean.dtypes)
print()
print(cy_young_clean.head(6).to_string())
```

## Merge all 8 datasets into one Master dataset

### Prior to merging normalized names

```python
import unicodedata

def normalize_name(name):
    """Remove accents and special characters from names"""
    return unicodedata.normalize('NFKD', str(name)).encode('ascii', 'ignore').decode('ascii')

# Apply to all dataframes
fg_stats_clean['Name'] = fg_stats_clean['Name'].apply(normalize_name)
fg_pitch_type_clean['Name'] = fg_pitch_type_clean['Name'].apply(normalize_name)
fg_velocity_clean['Name'] = fg_velocity_clean['Name'].apply(normalize_name)
fg_h_movement_clean['Name'] = fg_h_movement_clean['Name'].apply(normalize_name)
fg_v_movement_clean['Name'] = fg_v_movement_clean['Name'].apply(normalize_name)
fg_spin_clean['Name'] = fg_spin_clean['Name'].apply(normalize_name)
fg_advanced_clean['Name'] = fg_advanced_clean['Name'].apply(normalize_name)
cy_young_clean['Name'] = cy_young_clean['Name'].apply(normalize_name)
```

### Fixed 18 edge cases where merge failed

```python
# Manual name fixes for fg_advanced
name_fixes = {
    'Jon Niese': 'Jonathon Niese',
    'Nathan Karns': 'Nate Karns',
    'Rubby de la Rosa': 'Rubby De La Rosa',
    'Hyun-Jin Ryu': 'Hyun Jin Ryu',
    'Kwang-hyun Kim': 'Kwang Hyun Kim',
    'Glenn Otto': 'Glenn Otto Jr.',
    'Louie Varland': 'Louis Varland',
    'Luis L. Ortiz': 'Luis Ortiz',
    'Cameron Schlittler': 'Cam Schlittler'
}

fg_advanced_clean['Name'] = fg_advanced_clean['Name'].replace(name_fixes)

# Verify
in_stats = set(zip(fg_stats_clean['Season'], fg_stats_clean['Name']))
in_advanced = set(zip(fg_advanced_clean['Season'], fg_advanced_clean['Name']))
only_in_stats = in_stats - in_advanced

print(f'Remaining mismatches: {len(only_in_stats)}')
```

### Merged fangraph data

```python
# Drop vFA from fg_stats since fg_velocity already has it
fg_stats_clean = fg_stats_clean.drop(columns=['vFA'])

# Fix K%, BB%, K-BB% string formatting in fg_advanced
fg_advanced_clean['K%'] = fg_advanced_clean['K%'].str.replace('%', '').astype(float) / 100
fg_advanced_clean['BB%'] = fg_advanced_clean['BB%'].str.replace('%', '').astype(float) / 100
fg_advanced_clean['K-BB%'] = fg_advanced_clean['K-BB%'].str.replace('%', '').astype(float) / 100

print('fg_stats columns:', list(fg_stats_clean.columns))
print('K% sample:', fg_advanced_clean['K%'].head(3).values)
```

```python
# Step 1 - Merge all FanGraphs files on Season + Name
master = fg_stats_clean.merge(fg_pitch_type_clean[['Season', 'Name', 'FA%', 'FC%', 'FS%', 'FO%', 'SI%', 'SL%', 'CU%', 'KC%', 'EP%', 'CH%', 'SC%', 'KN%']], 
                               on=['Season', 'Name'], how='left')

master = master.merge(fg_velocity_clean.drop(columns=['Team', 'IP']), 
                      on=['Season', 'Name'], how='left')

master = master.merge(fg_h_movement_clean.drop(columns=['Team', 'IP']), 
                      on=['Season', 'Name'], how='left')

master = master.merge(fg_v_movement_clean.drop(columns=['Team', 'IP']), 
                      on=['Season', 'Name'], how='left')

master = master.merge(fg_spin_clean.drop(columns=['Team', 'IP']), 
                      on=['Season', 'Name'], how='left')

master = master.merge(fg_advanced_clean.drop(columns=['Team']), 
                      on=['Season', 'Name'], how='left')

print('After FanGraphs merge:', master.shape)
print()
print(master.head(3).to_string())
```

```python
# Step 2 - Left join cy_young
master = master.merge(cy_young_clean.drop(columns=['Team', 'WAR']), 
                      on=['Season', 'Name'], how='left')

# Fill non-voters with 0
master['Share'] = master['Share'].fillna(0)
master['CY_winner'] = master['CY_winner'].fillna(0).astype(int)
master['Vote_Pts'] = master['Vote_Pts'].fillna(0).astype(int)
master['1st_Place'] = master['1st_Place'].fillna(0).astype(int)
master['Rank'] = master['Rank'].fillna(0).astype(int)

print('After cy_young merge:', master.shape)
print()
print(master.head(3).to_string())
```

```python
# Verify cy_young merge worked correctly
print('Total pitchers:', len(master))
print('Cy Young voters:', (master['Share'] > 0).sum())
print('Cy Young winners:', master['CY_winner'].sum())
print('Expected winners (22 AL + 22 NL = 22 years x 2):', 22)
print()
print('Null counts for key columns:')
print(master[['Share', 'CY_winner', 'Vote_Pts', 'ERA', 'WAR', 'FIP']].isnull().sum())
```

```python
# Check cy_young name mismatches
in_stats = set(zip(fg_stats_clean['Season'], fg_stats_clean['Name']))
in_cy = set(zip(cy_young_clean['Season'], cy_young_clean['Name']))

only_in_stats = in_stats - in_cy
only_in_cy = in_cy - in_stats

print(f'In cy_young but not in fg_stats: {len(only_in_cy)}')
print()
print('Names in cy_young but not fg_stats:')
for s, n in sorted(only_in_cy):
    print(f'  {s} — "{n}"')
```

```python
# Save master dataset
master.to_csv('../Data/Processed/master_dataset.csv', index=False)
print('Saved master_dataset.csv with shape:', master.shape)
```
